**Loading Megaset & Visualization Setup**

```python
megaset = pd.read_csv('megaset.csv.zip', compression='zip')
wi_gdp = pd.read_csv('WI2022_GDP.csv')
```

```python
aggregate = {'share_total_obligated' : 'sum', 'share_total_outlayed' : 'sum', 'share_total_outlayed_capita' : 'sum', 'state_share_obligated':'sum', 'state_share_outlayed': 'sum',
             'total_obligated':'sum', 'fed_obligation':'sum', 'total_outlayed':'sum', 'population' : 'mean', 'total_outlayed_capita' : 'sum',
             'current_value': 'sum', 'base_and_exercised_options_value': 'sum', 'base_and_all_options_value': 'sum', 
             'potential_value': 'sum', 'fiscal_year' :'mean'}
```

```python
def billions(x, pos):
    return '{:,.0f}'.format(x / 1000000000) + 'B'

def millions(x, pos):
    return '{:,.0f}'.format(x / 1000000) + 'M'

def thousands(x, pos):
    return '{:,.0f}'.format(x / 1000) + 'K'
```

```python
fig, ax = plt.subplots()

sns.barplot(data = wi_gdp, x = 'GDP', y = 'Sector', palette = wi_colors, saturation=1)
ax.xaxis.set_major_formatter(FuncFormatter(billions))

#ax.set_title("Wisconsin Economy by Sector (2022)")
ax.set_ylabel("")

for p in ax.patches:
    width = p.get_width()
    ax.annotate('{:,.1f}'.format(width / 1000000000), 
                (width, p.get_y() + p.get_height() / 2.), 
                ha='center', va='center', 
                xytext=(15, 0), 
                textcoords='offset points')
    
sns.despine()
plt.savefig('figure1.png', bbox_inches='tight')
plt.show()
```

**Figure4: Share of New Federal Contract Obligations per Capita in the Midwest by State**

```python
state_colors = ["#EBB434", "#379A8B", "#006BA2", "#3EBCD2", "#C5050C", "#B4BA39"]
```

```python
pivoted_data = all_by_year.pivot_table(index='fiscal_year', columns='state_code', values='share_total_obligated_capita', aggfunc='sum')


desired_order = ['MN', 'IN', 'MI', 'IA', 'WI', 'IL']
pivoted_data = pivoted_data[desired_order]

# Convert to percentage if needed
pivoted_data_percentage = pivoted_data * 100
fig, ax = plt.subplots(figsize=(10, 5))

# Plotting each state's share as a part of the stacked bar
pivoted_data_percentage.plot(kind='barh', stacked=True, ax=ax, color=state_colors)

# Setting labels and title
plt.xlabel('Percentage')
plt.ylabel('')

#plt.ylabel('Fiscal Year')
#plt.title("Share of New Federal Contract Obligations per Capita in the Midwest", fontsize = 13)

# Improve layout
plt.tight_layout()

# Despine
sns.despine()

# Overlay percentages over the bars
for spine in ax.spines.values():
    spine.set_visible(False)

# Overlay percentages over the bars
for bar in ax.patches:
    width = bar.get_width()
    label_text = f'{width:.0f}%'  # Format the percentage
    if width > 0:  # If the bar is empty, don't place the text
        x = bar.get_x() + bar.get_width() / 2
        y = bar.get_y() + bar.get_height() / 2
        ax.text(x, y, label_text, ha='center', va='center', color='black')

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1], frameon=False, title='State Code', bbox_to_anchor=(1.00, 1), loc='upper left')

plt.tight_layout()

plt.savefig('figure4.png', bbox_inches='tight')

plt.show()
```
